```python
def gradient_descent(x,y):
    w_current=b_current=0
    iterations=1000
    m=len(x)
    learning_rate=0.0003
    
    for i in range(iterations):
        
        y_pred = w_current * x + b_current
        
        cost=(1/m)*sum([val**2 for val in (y-y_pred)])
        
        w_diff= -(1/m)*sum(x*(y-y_pred))
        b_diff= -(1/m)*sum(y-y_pred)
        
        w_current=w_current - learning_rate * w_diff
        b_current=b_current - learning_rate*b_diff
        
        print("w={}\t ,b={}\t ,iteration={}\t,costfunction={}".format(w_current,b_current,i,cost))
        
        
gradient_descent(x,y)
```
