```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
import numpy as np
```

```python
digits = load_digits()
```

```python
dir(digits)
#dir function return all the methods and attributes associated with the object.
```

```python
digits.data[0]
```

It will give an array ,  as images are stored inform of vectores of colour codes.

```python
plt.gray()

for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[:5]
```

This is the target variable, the output.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
model.predict([digits.data[67]])
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test , y_predicted)
cm
```

```python
import seaborn as sn


plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, annot_kws={"size": 12}, cmap='Blues')  # You can adjust the size and color
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
