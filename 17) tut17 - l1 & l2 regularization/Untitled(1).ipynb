```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Melbourne_housing_FULL.csv")
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[cols_to_use]
```

```python
df
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
cols_to_fill_mean = ['Landsize', 'BuildingArea']
df[cols_to_fill_mean] = df[cols_to_fill_mean].fillna(df[cols_to_fill_mean].mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
y = df.Price
```

```python
df.drop(df[['Price']], axis=1, inplace = True)
```

```python
df
```

```python
y.isnull().sum()
```

```python
y=y.dropna()
```

```python
df = pd.get_dummies(df, dtype='int', drop_first=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df,y,test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
model.score(xtrain,ytrain)
```

```python
from sklearn.linear_model import Lasso #L1 Reg
lassso = Lasso()
lassso.fit(xtrain,ytrain)
lassso.score(xtest,ytest)
```

```python
lassso.score(xtrain,ytrain)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(xtrain,ytrain)
```

```python
ridge.score(xtest,ytest)
```

```python
ridge.score(xtrain,ytrain)
```
