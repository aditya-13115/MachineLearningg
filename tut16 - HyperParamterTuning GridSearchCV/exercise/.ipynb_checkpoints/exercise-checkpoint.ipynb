````python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_digits
````

````python
digits = load_digits()
````

````python
dir(digits)
````

````python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
````

````python
y = digits.target
````

````python
y
````

````python
df
````

````python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB , MultinomialNB
from sklearn.tree import DecisionTreeClassifier
````

````python
models = {
    "SVC" : {
        "model" : SVC(gamma='auto'),
        "params" : {
            "C" : [1,10,20,30,40,50],
            'kernel' : ['linear', 'poly', 'rbf']
        }        
            },
    
    "RandomForestClassifier" : {
        "model" : RandomForestClassifier(criterion='gini'),
        "params" : {
            "n_estimators" : [1,10,100,200],
        }
            },
    
     "LogisticRegression" : {
        "model" : LogisticRegression(),
        "params" : {
            "penalty" : ['l1','l2'],
            "max_iter" : [10,100,200,300],
        }
            },

    "GaussianNB" : {
        "model" : GaussianNB(),
        "params" : {
            
        }
            },

    
    "MultinomialNB" : {
        "model" : MultinomialNB(),
        "params" : {
            'alpha' : [1,2,3],
        }
            },


    
    "DecisionTreeClassifier" : {
        "model" : DecisionTreeClassifier(),
        "params" : {
            'criterion' : ["gini", "entropy"],
            'splitter' : ["best", "random"],
            
        }
            },
    
}
````

````python
from sklearn.model_selection import GridSearchCV
````

````python
scores = []

for mName, mPara in models.items():
    clf = GridSearchCV(mPara['model'],mPara['params'],cv=5,return_train_score=False)
    clf.fit(df,y)
    scores.append({
            'model' : mName,
            'best_score' : clf.best_score_,
            'best_para' : clf.best_params_
    })

    dfScore = pd.DataFrame(scores,columns=['model','best_score','best_para'])
    dfScore
````
