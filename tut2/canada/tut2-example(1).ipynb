```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df
```

```python
#renaming the column "per capita income" to "pci"
df=df.rename(columns={'per capita income (US$)' : 'pci'})
                    
```

```python
df
```

```python
plt.scatter(df.year,df.pci,color= 'red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (in $)")
plt.title("Year vs Income")
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
reg.predict([[1974]])
```

```python
input = df
```

```python
input.drop(columns=['pci'])
```

```python
X = df[['year']]  # Use only the year column for training
```

```python
prediction=reg.predict(X)
```

```python
input['prediction']= prediction
```

```python
input
```

```python
plt.scatter(df.year,df.pci,color='red', marker='+')
plt.plot(input.year,reg.predict(input[['year']]),color='blue')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (in $)")
plt.title("Year vs Income [prediciton]")
plt.show()
```

```python
output=input.to_csv('output.csv',index=False)
```
