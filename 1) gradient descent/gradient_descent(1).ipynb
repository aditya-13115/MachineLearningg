```python
def gradient_descent(x,y):
    w_current, b_current = 0, 0  # Initialize weights
    iterations = 10000  # Number of iterations
    learning_rate = 0.08  # Learning rate
    n = len(x)  # Number of data points

    for i in range(iterations):  # Iterate over the number of iterations
        y_pred = (w_current * x) + b_current  # Predicted values
        costFunction = (1/n) * sum((y - y_pred) ** 2)  # Mean Squared Error
        
        # Compute gradients
        wd = -(2/n) * sum(x * (y - y_pred))  
        bd = -(2/n) * sum(y - y_pred)

        # Update weights
        w_current = w_current - (learning_rate * wd)
        b_current = b_current - (learning_rate * bd)

        # Print progress
        print(f'w: {w_current}, b: {b_current}, costFunction: {costFunction}, iteration: {i}.')

```

```python
gradient_descent(x,y)
```

```python
import numpy as np

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Logistic Regression using Gradient Descent
def logistic_regression(X, y, learning_rate=0.01, iterations=1000):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)  # Initialize weights
    bias = 0  # Initialize bias

    for i in range(iterations):
        # Compute predictions
        linear_model = np.dot(X, weights) + bias
        y_pred = sigmoid(linear_model)

        # Compute gradients
        dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
        db = (1 / n_samples) * np.sum(y_pred - y)

        # Update weights and bias
        weights -= learning_rate * dw
        bias -= learning_rate * db

        # Print cost every 100 iterations
        if i % 100 == 0:
            cost = -(1 / n_samples) * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
            print(f"Iteration {i}: Cost = {cost:.4f}")

    return weights, bias

# Sample dataset (binary classification)
X = np.array([[1], [2], [3], [4], [5]])  # Feature
y = np.array([0, 0, 1, 1, 1])  # Labels (0 or 1)

# Train the model
w_final, b_final = logistic_regression(X, y)

# Prediction function
def predict(X_new):
    linear_model = np.dot(X_new, w_final) + b_final
    probabilities = sigmoid(linear_model)
    return [1 if p > 0.5 else 0 for p in probabilities]

# Test prediction
X_test = np.array([[1.5], [3.5], [4.5]])
predictions = predict(X_test)
print("Predictions:", predictions)
```
